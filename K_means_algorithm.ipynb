```python
#Start by visualizing some data points:
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/abcde.csv')
df.head()
```

```python
plt.scatter(df.rollno,df['marks'])
plt.xlabel('rollno')
plt.ylabel('marks')
```

```python
km = KMeans(n_clusters=3)
predicted = km.fit_predict(df[['rollno','marks']])
predicted
```

```python
df['cluster']=predicted
df.head()
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.rollno,df1['marks'],color='green')
plt.scatter(df2.rollno,df2['marks'],color='red')
plt.scatter(df3.rollno,df3['marks'],color='blue')
plt.xlabel('rollno')
plt.ylabel('marks')
```

```python
scale = MinMaxScaler()

scale.fit(df[['marks']])
df['marks'] = scale.transform(df[['marks']])

scale.fit(df[['rollno']])
df['rollno'] = scale.transform(df[['rollno']])

km = KMeans(n_clusters=3)
predicted = km.fit_predict(df[['rollno','marks']])
predicted
```

```python
df = df.drop(['cluster'], axis='columns')

df['cluster']=predicted
df.head()

df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.rollno,df1['marks'],color='green')
plt.scatter(df2.rollno,df2['marks'],color='red')
plt.scatter(df3.rollno,df3['marks'],color='blue')
plt.xlabel('rollno')
plt.ylabel('marks')
```

```python
km.cluster_centers_
```

```python
plt.scatter(df1.rollno,df1['marks'],color='green')
plt.scatter(df2.rollno,df2['marks'],color='red')
plt.scatter(df3.rollno,df3['marks'],color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*')
plt.xlabel('rollno')
plt.ylabel('marks')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your data from a CSV file
daf = pd.read_csv('/content/drive/MyDrive/abcde.csv')
x1 = daf['rollno']
x2 = daf['marks']

def distance(x1, x2, m1, m2, n1, n2):
    c1 = []
    c2 = []
    for i in range(len(x1)):
        d1 = np.sqrt(((x1[i] - m1) ** 2) + ((x2[i] - m2) ** 2))
        d2 = np.sqrt(((x1[i] - n1) ** 2) + ((x2[i] - n2) ** 2))
        if d1 <= d2:
            c1.append(i)
        else:
            c2.append(i)
    return c1, c2

def new_centroid(x1, x2, c1, c2):
    m1 = m2 = n1 = n2 = 0
    for i in range(len(c1)):
        m1 += x1[c1[i]]
        m2 += x2[c1[i]]
    for i in range(len(c2)):
        n1 += x1[c2[i]]
        n2 += x2[c2[i]]
    m1 = m1 / len(c1)
    m2 = m2 / len(c1)
    n1 = n1 / len(c2)
    n2 = n2 / len(c2)
    return m1, m2, n1, n2

# Initialize centroids
m1 = x1[0]
m2 = x1[1]
n1 = x2[0]
n2 = x2[1]

# Number of iterations for K-means
num_iterations = 3

for i in range(num_iterations):
    c1, c2 = distance(x1, x2, m1, m2, n1, n2)
    m1, m2, n1, n2 = new_centroid(x1, x2, c1, c2)
    print("Cluster 1 Centroid: (", m1, ",", m2, ")")
    print("Cluster 2 Centroid: (", n1, ",", n2, ")")

# Plot the clusters
plt.scatter(x1[c1], x2[c1], label='Cluster 1', c='red')
plt.scatter(x1[c2], x2[c2], label='Cluster 2', c='blue')

plt.scatter(m1, m2, c='black', marker='x', s=100, label='Cluster 1 Centroid')
plt.scatter(n1, n2, c='green', marker='x', s=100, label='Cluster 2 Centroid')

plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.legend()
plt.show()

```
