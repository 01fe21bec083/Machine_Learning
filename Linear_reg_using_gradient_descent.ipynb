## 1. Gradient Descent for (single variable ) Linear Regression

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# Load the data
k = pd.read_excel('/content/data.xlsx', usecols='A')
l = pd.read_excel('/content/data.xlsx', usecols='B')
x = np.array(k).reshape((-1, 1))
y = np.array(l)
x_min = np.min(x)
x_max = np.max(x)
y_min = np.min(y)
y_max = np.max(y)
x = (x - x_min) / (x_max - x_min)
y = (y - y_min) / (y_max - y_min)

from sklearn.linear_model import LinearRegression

SLR = LinearRegression()
SLR.fit(x, y)
pred1 = SLR.predict(x)
print(pred1)

Wo = 0.0  # Initialize with 0
W1 = 0.0  # Initialize with 0
alph = float(input("Enter the learning rate: "))
m = len(x)  # The number of data points

def gradient(Wo, W1, x, y, alpha):
    for i in range(1000):
        mse1 = 0
        mse2 = 0
        cost = 0.0

        for j in range(m):
            pred = Wo + W1 * x[j]
            mse1 += pred - y[j]
            mse2 += (pred - y[j]) * x[j]
            cost += (pred - y[j]) ** 2

        Wo -= (alpha / m) * mse1
        W1 -= (alpha / m) * mse2

        cost /= (2.0 * m)

    return Wo, W1

Wo, W1 = gradient(Wo, W1, x, y, alph)
pred = Wo + W1 * x

print(pred)

plt.scatter(x, y, color='Red')
plt.plot(x, pred, color='Black', marker='*', linestyle='dotted')
plt.show()

```
