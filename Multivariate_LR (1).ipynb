# Multivariate Linear Regression
1. the excel file should have multiple features( columns)
2. the correlation between features and Y. (Denoted by R)


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Read the data
k = pd.read_excel('/content/drive/MyDrive/multivariate_example2.xlsx', usecols='C,D,E')
l = pd.read_excel('/content/drive/MyDrive/multivariate_example2.xlsx', usecols='F')

# Convert data to arrays
x = np.array(k).reshape((-1,3))
y = np.array(l)

# Normalize the features (x)
mean = np.mean(x, axis=0)
std = np.std(x, axis=0)
x_nor = (x - mean) / std

# Your gradient descent implementation
Wo = float(input("Enter the bias value"))
W1 = float(input("Enter the W1 value"))
W2 = float(input("Enter the W2 value"))
W3 = float(input("Enter the W3 value"))
alph = float(input("Enter the learning rate"))
n = int(input("Enter the number of iterations"))
m = len(x_nor)

for i in range(n):
    mse1 = 0
    mse2 = 0
    mse3 = 0
    mse4 = 0
    cost = 0.0
    for j in range(m):
        pred = Wo + W1 * x_nor[j][0] + W2 * x_nor[j][1] + W3 * x_nor[j][2]
        mse1 += pred - y[j]
        mse2 += (pred - y[j]) * x_nor[j][0]
        mse3 += (pred - y[j]) * x_nor[j][1]
        mse4 += (pred - y[j]) * x_nor[j][2]
        cost += (pred - y[j]) ** 2

    Wo = Wo - (alph / m) * mse1
    W1 = W1 - (alph / m) * mse2
    W2 = W2 - (alph / m) * mse3
    W3 = W3 - (alph / m) * mse4

    cost /= (2.0 * m)

# Predictions using your gradient descent model
pred1 = Wo + W1 * x_nor[:, 0] + W2 * x_nor[:, 1] + W3 * x_nor[:, 2]

# Linear regression using scikit-learn
MLR = LinearRegression()
MLR.fit(x_nor, y)
pred2 = MLR.predict(x_nor)
print(pred1)
print(pred2)
```
