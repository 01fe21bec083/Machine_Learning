# Multivariate Linear Regression
1. the excel file should have multiple features( columns)
2. the correlation between features and Y. (Denoted by R)


```python
#Task1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
k=pd.read_excel('/content/multivariate_example2.xlsx',usecols='C,D,E')
l=pd.read_excel('/content/multivariate_example2.xlsx',usecols='F')

x=np.array(k).reshape((-1,3))
y=np.array(l)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
MLR=LinearRegression()#creating an instance as Multivariate linear regression
```

```python
MLR.fit(x,y)
```

```python
pred=MLR.predict(x)
print(pred)
```

```python
pred=MLR.intercept_+np.sum(MLR.coef_*x,axis=1)
print(pred)
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Black',marker='*')

print("MSE:",mse)
print("Intercept=",MLR.intercept_)
print("Slope=",MLR.coef_)
```

```python
predCO2=MLR.predict([[3300, 1300,500]])
print(predCO2)
```
