## 1) string palindrome check

```python
#1
n=input("Enter the string: ")
count=0
j=-1
for i in range((len(n))//2):
  if n[i]==n[j]:
    j-=1
  else:
    count=-1
    break
if count==0:
  print("string is palindrome")
else:
  print("String is not palindrome")
```

# 2) star pattern

```python
#2
n=int(input("Enter the numbe rof rows"))
for i in range(n):
  for j in range(n):
    if i>=j:
      print("*",end=" ")
  print("\n")
```

# 3)Odd and Even print

```python
#3
lis=[]
odd=[]
even=[]
for i in range(10):
  x=int(input("Enter the element"))
  lis.append(x)
for i in range(10):
  if lis[i]%2==0:
    even.append(lis[i])
  else:
    odd.append(lis[i])
print("Odd list is ",odd)
print("Even list is ",even)

```

# 4) Prime or not prime

```python
#4
def prime_check(x):
  count=0
  for i in range(1,x+1):
    if x%i==0:
      count+=1
  if count==2:
    return 1
  else:
    return -1
n=int(input("Enter the number "))
y=prime_check(n)
if y==1:
 print("It is prime number")
else:
 if n%3==0:
   print("It is divisible by 3")
 elif n%7==0:
  print("It is divisible by 7")
 elif n%9==0:
  print("It is divisible by 9")
 elif n%11==0:
  print("It is divisible by 11")
 else:
  print("It is not a prime number")
```
